##Laços e Repetições

```python
def verificar_se_pode_dirigir(idade):
  if idade >= 18:
    print(f'{idade} anos de idade, TEM permissão para dirigir')
  else:
    print(f'{idade} anos de idade, NÃO TEM permissão para dirigir')
```

```python
verificar_se_pode_dirigir(30)
```

```python
idades = [15,18,20,16,30]
for item in idades: #Para cada 'item' dentro da lista idades
  verificar_se_pode_dirigir(item)
```

# Atividade Individual Prática

Modelo de Random Forest *site: e-commerce*

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
uri = "https://raw.githubusercontent.com/daniel-usp/LIT/master/Social_Network_Ads.csv"
df = pd.read_csv(uri)
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
x = df.drop('Purchased', axis=1)
y = df['Purchased']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state = 101)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=100, criterion = 'gini', max_depth=10, n_estimators=50, n_jobs=1)
rfc.fit(x_train, y_train)

features = x_test.columns
features_importance = zip(rfc.feature_importances_,features)
for importance, feature in sorted(features_importance, reverse = True):
  print("%s: %f%%" % (feature, importance*100))
```

```python
previsoes = rfc.predict(x_test)
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, previsoes))
```
